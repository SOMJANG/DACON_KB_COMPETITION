```
data2 = pd.read_csv("model_kb_dacon02_trainset.csv")
train_data_text = list(data2['clear_text2'])

train_clear_text_len2 = []

for i in range(len(train_data_text)):
  train_clear_text_len2.append(len(train_data_text[i]))
data2['len3'] = train_clear_text_len2
data2 = data2.sort_values(by=['len3'], axis=0, ascending=False)
data2
```
